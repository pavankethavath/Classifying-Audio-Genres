```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier, XGBRegressor  # Advanced ML model
```

```python
# Select only numerical columns
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Plot distributions for each numerical column
plt.figure(figsize=(40, 50))  # Increase figure size for better spacing
for i, col in enumerate(numeric_columns, 1):
    plt.subplot((len(numeric_columns) // 3) + 1, 3, i)
    sns.histplot(data[col], kde=True, log_scale=(False, True) if data[col].skew() > 1 else False)  # Log scale if highly skewed
    plt.title(f'Distribution of {col}', fontsize=30)
    plt.xlabel(col, fontsize=25)
    plt.ylabel('Frequency', fontsize=25)
    plt.xticks(fontsize=18)  # Increase x-axis tick label size
    plt.yticks(fontsize=18)  # Increase y-axis tick label size
    
plt.tight_layout()
plt.show()
```

```python
# Boxplot for each numeric feature
plt.figure(figsize=(15, 30))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(5, 3, i)
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
data['explicit'] = data['explicit'].apply(lambda x: 1 if x == 'true' else 0)
```

```python
# Check unique values in categorical columns
categorical_columns = data.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(f"{col}: {data[col].nunique()} unique values")
```

```python
data = data.drop(columns=['artists', 'album_name', 'track_name'])
```

```python
# Display distribution of target variable (e.g., genres or popularity)
plt.figure(figsize=(30,10))
sns.countplot(data['track_genre'])
plt.title('Distribution of Track Genres')
plt.xticks(rotation=90)
plt.show()
```

```python
genre_encoder = LabelEncoder()
data['track_genre'] = genre_encoder.fit_transform(data['track_genre'])
```

```python
# Plot heatmap for the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Selecting numeric features for scaling
numeric_features = ['popularity', 'duration_ms', 'explicit','danceability', 'energy', 
                    'loudness', 'speechiness', 'acousticness', 
                    'instrumentalness', 'liveness', 'valence', 'tempo','time_signature']

scaler = StandardScaler()
data[numeric_features] = scaler.fit_transform(data[numeric_features])
```

```python
data.head()
```

```python
# Elbow method for optimal k
inertia = []
k_range = range(1, 15)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow Curve
plt.figure(figsize=(8, 4))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```
